```python
import os
import numpy as np
import joblib
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
from sklearn.ensemble import RandomForestClassifier

# ==========================================
# 1. DYNAMIC HEADER READER
# ==========================================
def read_envi_header(bil_path):
    """Automatically finds and reads the .hdr file to get correct dimensions."""
    hdr_path = bil_path.replace('.bil', '.hdr')
    dims = {'lines': 500, 'samples': 900, 'bands': 300} # Default fallback

    if not os.path.exists(hdr_path):
        print(f"⚠️ WARNING: Could not find {hdr_path}. Using default 500x900x300.")
        return dims['lines'], dims['samples'], dims['bands']

    print(f"📄 Reading header file: {hdr_path}")
    with open(hdr_path, 'r') as f:
        for line in f:
            line = line.lower()
            if 'lines' in line and '=' in line:
                dims['lines'] = int(line.split('=')[1].strip())
            elif 'samples' in line and '=' in line:
                dims['samples'] = int(line.split('=')[1].strip())
            elif 'bands' in line and '=' in line:
                dims['bands'] = int(line.split('=')[1].strip())

    print(f"✅ Dynamic Dimensions Found: {dims['lines']} Lines, {dims['samples']} Samples, {dims['bands']} Bands.")
    return dims['lines'], dims['samples'], dims['bands']

# ==========================================
# 2. FULL MODEL BLUEPRINT
# ==========================================
class RobustTournamentRF:
    def __init__(self, n_estimators=100):
        self.rf_stage1 = RandomForestClassifier(n_estimators=n_estimators, random_state=42, class_weight='balanced')
        self.rf_stage2 = RandomForestClassifier(n_estimators=n_estimators, random_state=42, class_weight='balanced')
        self.expected_bands = 0

    def fit(self, X, y):
        pass

    def _adjust_bands(self, X):
        current_bands = X.shape[1]
        if current_bands == self.expected_bands:
            return X
        elif current_bands < self.expected_bands:
            padded = np.zeros((X.shape[0], self.expected_bands), dtype=X.dtype)
            padded[:, :current_bands] = X
            return padded
        else:
            return X[:, :self.expected_bands]

    def predict(self, X):
        X_safe = self._adjust_bands(X)
        s1_preds = self.rf_stage1.predict(X_safe)
        s2_preds = self.rf_stage2.predict(X_safe)

        final_preds = np.zeros(len(X_safe), dtype=int)
        for i in range(len(X_safe)):
            if s1_preds[i] == 1:
                final_preds[i] = 1
            else:
                if s2_preds[i] == 0:
                    final_preds[i] = 0
                else:
                    final_preds[i] = 2
        return final_preds

# ==========================================
# 3. VISION & MASKING
# ==========================================
# ==========================================
# 3. VISION & MASKING (Upgraded Smart Mask)
# ==========================================
def extract_leaf(cube, initial_threshold=0.10):
    print(f"👁️ Masking out background (Initial NDVI threshold: {initial_threshold})...")
    data = (cube - cube.min()) / (cube.max() - cube.min() + 1e-6)

    # Calculate NDVI
    ndvi = (data[:, :, 200] - data[:, :, 140]) / (data[:, :, 200] + data[:, :, 140] + 1e-6)

    # Try initial threshold
    mask = ndvi > initial_threshold

    # FAILSAFE: Auto-lower the threshold if no leaf is found
    if not np.any(mask):
        print("⚠️ Strict mask failed. Auto-lowering threshold to 0.05...")
        mask = ndvi > 0.05

    if not np.any(mask):
        print("⚠️ Still no leaf found. Auto-lowering to 0.01...")
        mask = ndvi > 0.01

    if not np.any(mask):
        print("⚠️ Extreme fallback: Just keeping the brightest pixels...")
        # If NDVI completely fails, just take the top 20% brightest pixels overall
        brightness = np.mean(data, axis=2)
        mask = brightness > np.percentile(brightness, 80)

    return mask

# ==========================================
# 4. FULL DYNAMIC LEAF SCANNER
# ==========================================
# ==========================================
# 4. FULL DYNAMIC LEAF SCANNER (Fixed True-Length)
# ==========================================
def run_full_diagnostic_pipeline(bil_path, health_model_path):
    print("🛰️ Loading hyperspectral data...")

    # 1. GET HEADER DIMS
    # We trust samples and bands, but we will ignore the header's "lines"
    _, ncols, nbands = read_envi_header(bil_path)

    # 2. LOAD RAW DATA
    raw_data = np.fromfile(bil_path, dtype=np.uint16)

    # 3. CALCULATE TRUE IMAGE LENGTH DYNAMICALLY
    # A scanner takes a full row (ncols * nbands) at a time.
    # Let's see how many full rows actually exist in your file!
    elements_per_row = ncols * nbands
    nrows = len(raw_data) // elements_per_row

    print(f"📏 True image length calculated from file size: {nrows} lines!")

    # Drop any tiny incomplete data fraction at the very end of the file
    valid_elements = nrows * elements_per_row
    raw_data = raw_data[:valid_elements]

    # 4. RESHAPE (Using the true, un-cut nrows)
    cube = raw_data.reshape((nrows, nbands, ncols))
    cube = np.transpose(cube, (0, 2, 1)) # Now shape is (nrows, ncols, nbands)

    # 5. FIND THE LEAF
    leaf_mask = extract_leaf(cube)
    flat_mask = leaf_mask.flatten()
    leaf_indices = np.where(flat_mask)[0]

    if len(leaf_indices) == 0:
        print("❌ ERROR: No leaf pixels found! The background mask removed everything.")
        return

    # 6. LOAD AI & PREDICT
    print("🧠 Loading Unified AI Model...")
    model = joblib.load(health_model_path)

    flat_cube = cube.reshape(-1, nbands).astype(np.float32)
    leaf_pixels = flat_cube[leaf_indices]

    print(f"🔬 Analyzing {len(leaf_indices):,} individual leaf pixels...")
    predictions = model.predict(leaf_pixels)

    # 7. REBUILD IMAGE OVERLAY
    heatmap = np.full((nrows, ncols), -1, dtype=int)
    heatmap.flat[leaf_indices] = predictions

    # 8. CALCULATE STATS
    unique, counts = np.unique(heatmap, return_counts=True)
    stats = dict(zip(unique, counts))

    total_leaf = stats.get(0, 0) + stats.get(1, 0) + stats.get(2, 0)
    p_healthy = (stats.get(0, 0) / total_leaf * 100) if total_leaf else 0
    p_mid = (stats.get(1, 0) / total_leaf * 100) if total_leaf else 0
    p_severe = (stats.get(2, 0) / total_leaf * 100) if total_leaf else 0

    # 9. PLOT AND PRINT
    print("🎨 Rendering Diagnostic Heatmap...")
    plt.figure(figsize=(12, 7))
    cmap = mcolors.ListedColormap(['black', '#228B22', '#FFD700', '#FF0000'])
    norm = mcolors.BoundaryNorm([-1.5, -0.5, 0.5, 1.5, 2.5], cmap.N)

    plt.imshow(heatmap, cmap=cmap, norm=norm)
    cbar = plt.colorbar(ticks=[-1, 0, 1, 2], fraction=0.03, pad=0.04)
    cbar.ax.set_yticklabels(['Background', 'Healthy (Green)', 'Midstage (Yellow)', 'Unhealthy (Red)'])
    plt.title("Whole Leaf AI Diagnostic Map", fontsize=16)
    plt.axis('off')
    plt.show()

    print("\n" + "="*40)
    print("📊 WHOLE LEAF HEALTH REPORT")
    print("="*40)
    print(f"🟢 Healthy:   {stats.get(0, 0):,} pixels ({p_healthy:.1f}%)")
    print(f"🟡 Midstage:  {stats.get(1, 0):,} pixels ({p_mid:.1f}%)")
    print(f"🔴 Unhealthy: {stats.get(2, 0):,} pixels ({p_severe:.1f}%)")
    print("="*40)

# ==========================================
# EXECUTE THE SCAN
# ==========================================
# Ensure the .hdr file is in the same folder as the .bil file!
run_full_diagnostic_pipeline(
    bil_path='/content/soybean_14thSept_Healthy.bil',
    health_model_path='/content/final_leaf_model.pkl'
)
run_full_diagnostic_pipeline(
    bil_path='/content/Hyperspectral.bil',
    health_model_path='/content/final_leaf_model.pkl'
)
```

```python
import os
import numpy as np
import joblib
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

# For the LLM Agronomist Agent
try:
    import google.generativeai as genai
    from google.colab import userdata
    HAS_LLM = True
except ImportError:
    HAS_LLM = False
    print("google.generativeai not installed. Agronomist will use basic rules.")


# 1. UTILITIES & DATA LOADING

def read_envi_header(bil_path):
    """Reads the .hdr file to get width (samples) and depth (bands)."""
    hdr_path = bil_path.replace('.bil', '.hdr')
    dims = {'samples': 900, 'bands': 300}

    if os.path.exists(hdr_path):
        with open(hdr_path, 'r') as f:
            for line in f:
                line = line.lower()
                if 'samples' in line and '=' in line:
                    dims['samples'] = int(line.split('=')[1].strip())
                elif 'bands' in line and '=' in line:
                    dims['bands'] = int(line.split('=')[1].strip())
    return dims['samples'], dims['bands']


# 2. VISION AGENT (Smart Masking)

def extract_leaf(cube, initial_threshold=0.10):
    print(f"👁️ Vision Agent: Masking out background...")
    data = (cube - cube.min()) / (cube.max() - cube.min() + 1e-6)

    ndvi = (data[:, :, 200] - data[:, :, 140]) / (data[:, :, 200] + data[:, :, 140] + 1e-6)
    mask = ndvi > initial_threshold

    # Auto-adjust if strict mask fails
    if not np.any(mask):
        mask = ndvi > 0.05
    if not np.any(mask):
        mask = ndvi > 0.01
    if not np.any(mask):
        brightness = np.mean(data, axis=2)
        mask = brightness > np.percentile(brightness, 80)

    return mask


# 3. EXPERT AI AGENTS

class BotanistAgent:
    def identify_plant(self, leaf_pixels):
        print("🌿 Botanist Agent: Analyzing spectral signature...")
        return "Soybean (Glycine max)" # Placeholder for future species .pkl

class AgronomistAgent:
    def __init__(self):
        self.model = None
        if HAS_LLM:
            try:
                print("🔌 Agronomist Agent: Connecting to Gemini AI...")
                api_key = userdata.get('GEMINI_API_KEY')
                genai.configure(api_key=api_key)
                self.model = genai.GenerativeModel('gemini-2.5-flash')
            except Exception:
                print("⚠️ Agronomist Agent: No GEMINI_API_KEY found in secrets. Using basic rules.")

    def generate_report(self, species, p_healthy, p_mid, p_severe):
        print("🧑‍🌾 Agronomist Agent: Formulating field recommendations...")

        # Use LLM if connected
        if self.model:
            prompt = f"""
            You are an expert Agronomist. I ran a hyperspectral scan on a {species} leaf.
            Results: Healthy: {p_healthy:.1f}%, Midstage Stress: {p_mid:.1f}%, Severe Damage: {p_severe:.1f}%.
            Write a short, professional crop intelligence report.
            Include: 1. Threat level. 2. Probable causes for {species}. 3. 2-3 immediate, actionable recommendations.
            Keep it concise with bullet points.
            """
            try:
                return self.model.generate_content(prompt).text
            except Exception as e:
                return f"❌ LLM Error: {str(e)}\nFalling back to standard report..."

        # Fallback Rule-Based Logic
        report = f"Target Crop: {species}\n"
        if p_severe > 15.0:
            report += "🚨 CRITICAL ALERT: Severe tissue damage. High probability of advanced necrosis or blight."
        elif p_mid > 20.0 or p_severe > 5.0:
            report += "⚠️ WARNING: Elevated stress levels. Check soil moisture and nitrogen levels."
        elif p_healthy > 85.0:
            report += "✅ OPTIMAL: Crop is healthy. Maintain current regimen."
        else:
            report += "ℹ️ NOTICE: Moderate stress detected. Continue monitoring."
        return report


# 4. MASTER PIPELINE

def run_expert_system(bil_path, health_model_path):
    print("\n" + "="*60)
    print(f"🚀 INITIATING SCAN: {os.path.basename(bil_path)}")
    print("="*60)

    # 1. LOAD & FIX SHAPE
    ncols, nbands = read_envi_header(bil_path)
    raw_data = np.fromfile(bil_path, dtype=np.uint16)

    elements_per_row = ncols * nbands
    nrows = len(raw_data) // elements_per_row
    valid_elements = nrows * elements_per_row
    raw_data = raw_data[:valid_elements]

    cube = raw_data.reshape((nrows, nbands, ncols))
    cube = np.transpose(cube, (0, 2, 1))

    # 2. VISION MASKING
    leaf_mask = extract_leaf(cube)
    flat_mask = leaf_mask.flatten()
    leaf_indices = np.where(flat_mask)[0]

    if len(leaf_indices) == 0:
        print("❌ ERROR: Vision Agent failed to find leaf tissue.")
        return

    flat_cube = cube.reshape(-1, nbands).astype(np.float32)
    leaf_pixels = flat_cube[leaf_indices]

    # 3. BOTANIST
    botanist = BotanistAgent()
    species = botanist.identify_plant(leaf_pixels)

    # 4. PATHOLOGIST (Your .pkl model)
    print("🧠 Pathologist Agent: Loading AI Model...")
    model = joblib.load(health_model_path)
    print(f"🔬 Pathologist Agent: Diagnosing {len(leaf_indices):,} pixels...")
    predictions = model.predict(leaf_pixels)

    # 5. STATS CALCULATION
    heatmap = np.full((nrows, ncols), -1, dtype=int)
    heatmap.flat[leaf_indices] = predictions

    unique, counts = np.unique(heatmap, return_counts=True)
    stats = dict(zip(unique, counts))

    total_leaf = stats.get(0, 0) + stats.get(1, 0) + stats.get(2, 0)
    p_healthy = (stats.get(0, 0) / total_leaf * 100) if total_leaf else 0
    p_mid = (stats.get(1, 0) / total_leaf * 100) if total_leaf else 0
    p_severe = (stats.get(2, 0) / total_leaf * 100) if total_leaf else 0

    # 6. AGRONOMIST
    agronomist = AgronomistAgent()
    advice = agronomist.generate_report(species, p_healthy, p_mid, p_severe)

    # 7. RENDER DASHBOARD
    print("\n🎨 Generating Executive Dashboard...")
    plt.figure(figsize=(12, 7))
    cmap = mcolors.ListedColormap(['black', '#228B22', '#FFD700', '#FF0000'])
    norm = mcolors.BoundaryNorm([-1.5, -0.5, 0.5, 1.5, 2.5], cmap.N)

    plt.imshow(heatmap, cmap=cmap, norm=norm)
    cbar = plt.colorbar(ticks=[-1, 0, 1, 2], fraction=0.03, pad=0.04)
    cbar.ax.set_yticklabels(['Background', 'Healthy', 'Midstage', 'Unhealthy'])
    plt.title(f"Diagnostic Map: {species}\n{os.path.basename(bil_path)}", fontsize=16)
    plt.axis('off')
    plt.show()

    print("\n" + "="*60)
    print("📋 MULTI-AGENT EXPERT REPORT")
    print("="*60)
    print(f"🟢 Healthy Tissue:   {stats.get(0, 0):,} px ({p_healthy:.1f}%)")
    print(f"🟡 Stressed Tissue:  {stats.get(1, 0):,} px ({p_mid:.1f}%)")
    print(f"🔴 Necrotic Tissue:  {stats.get(2, 0):,} px ({p_severe:.1f}%)")
    print("-" * 60)
    print(advice)
    print("="*60 + "\n")


# 5. EXECUTE THE PIPELINE

run_expert_system('/content/soybean_14thSept_Healthy.bil', '/content/final_leaf_model.pkl')
run_expert_system('/content/Hyperspectral.bil', '/content/final_leaf_model.pkl')
```
